```python
def uni(df):

    for i in range(len(df.columns)):
        print('\n All Unique Value in ' + str(df.columns[i]))
        print(np.sort(df[df.columns[i]].unique()))
        print('Total no of unique values ' +
                str(len(df[df.columns[i]].unique())))
```

```python
count = df.sum()
count.sort_values(0, ascending = False, inplace=True)
count = count.to_frame().reset_index()
count = count.rename(columns = {'index': 'Books',0: 'count'})
count
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
rules=apriori(df,min_support=0.1,use_colnames=True)
rules
```

```python
results=list(rules)
```

```python
rules1=association_rules(rules,metric='lift',min_threshold=0.8)
rules1
```

```python
rules1.sort_values('lift',ascending=False)
```

```python
rules1[rules1.lift>1]
```

```python
plt.scatter(rules1['support'],rules1['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
